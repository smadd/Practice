```python
import pandas as pd
import numpy as np
```

```python
df = (pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|'))
users = df
users.head(10)
```

```python
print(users['age'].mean())
```

```python
dfoccupation =users.groupby('occupation').mean().age
dfoccupation
```

```python
occu_ratio = users.groupby(['occupation']).count()
print(occu_ratio)
```

```python
occu_ratio_gen = users.groupby(['occupation', 'gender']).count()
print(occu_ratio_gen)
```

```python
occu_ratio_gen['ratio'] = occu_ratio_gen['age'] *100 / occu_ratio['age']
occu_ratio_gen
```

```python
occu_ratio_gen_sort = occu_ratio_gen.sort_values(['ratio'], ascending=False)
occu_ratio_gen_sort
```

```python
male = occu_ratio_gen_sort.loc[pd.IndexSlice[:, ['M']], :]
male
```
